**CENTRAL TENDECY**
  - In statistics, a central tendency (or measure of central tendency) is a central or typical value for a probability distribution.
  - It may also be called *a center or location of the distribution.* 
  - The most common measures of central tendency are the arithmetic mean, the median, and the mode. 
  - A middle tendency can be calculated for either a finite set of values or for a theoretical distribution, such as the normal distribution.

1. **Mean**
  - Measure of Central Tendency
  - (Sum of data values)/(Data Count) 
  - AFFECTED BY OUTLIERS

2. **Median**
  - Also a Measure of Central Tendency
  - It is the observation/value at the middle when the data is sorted in ascending order
  - MIGHT OR MIGHT NOT AFFECT BY OUTLIERS

3. **Mode**
  - Most frequently observed variable (categorical column) in the column

4. **Quartile**
  - Firstly, Data is sorted in Ascending Order
    1. *First Quartile* = Lower Quartile = 25% of the data [0-25% of Data)
    2. *Second Quartile* = Median = 50% of the data (0-50% of Data)
    3. *Third Quartile* = Upper Quartile = 75% of the data (0-75% of Data)

5. **Percentile**
  - Nth percentile of an observation variable is the value that cuts off the first n percent of the data values when it is sorted in ascending order

6. **Range**
  - Measures how far apart the entire data spread is in terms of values 
  - Range = (Largest Value - Smallest Value)
  - AFFECTED BY OUTLIERS

7. **Inter Quartile Range**
  - IQR = (Upper Quartile - Lower Quartile) = (75% of Data - 25% of Data)

8. **Box Plot**
  - Graphical representation of
    1.  Three quantiles (First, Second and Third Quantile Respectively) 
    2.  Smallest value 
    3.  Largest value 

9. **Standard Deviation**
  - Square root of the Variance

10. **Variance**
  - Numerical Measure of how the data value is dispersed around the mean
  - AFFECTED BY OUTLIERS

11. **Covariance**
  - To check linearity between two variables. It can be +ve or -ve relationship.

12. **Correlation Coefficient**
  - Strength of relationship between two variables
      1. *Pearson Correlation Coefficient* - Works on linear relationship
      2. *Spearman Correlation Coefficient* - Works on both linear & non linear relationship

13. **Skewness**
  - Spreadness of the Distribution

14. **Kurtosis**
  - Peakness of the Distribution

- The median is the middle number in a sorted, ascending or descending, list of numbers and can be more descriptive of that data set than the average.
- The median is sometimes used as opposed to the mean when there are *outliers in the sequence that might skew the average of the values.*

- If there is an *even amount of numbers* in the list, the middle pair must be determined, added together, and divided by two to find the median value.

**INTERQUARTILE RANGE(IQR)**
- Imagine all the data in a set as points on a number line. 
- Once plotted on that number line, the smallest data point and the biggest data point in the set of data create the boundaries (i.e. a *lower bound and an upper bound*) of an interval of space on the number line that contains all data points in the set. 
- The interquartile range (IQR) is the length of the middle 50% of that interval of space(25% of Data to 75% of Data).
--------------------------------------------------------
*What if I Have an Even Set of Numbers?*

*Example question: Find the IQR for the following data set: 3, 5, 7, 8, 9, 11, 15, 16, 20, 21.*

**Step 1: Put the numbers in order.**

3, 5, 7, 8, 9, 11, 15, 16, 20, 21.

**Step 2: Make a mark in the center of the data:**

3, 5, 7, 8, 9, | 11, 15, 16, 20, 21.

**Step 3: Place parentheses around the numbers above and below the mark you made in Step 2–it makes Q1 and Q3 easier to spot.**

(3, 5, 7, 8, 9) | (11, 15, 16, 20, 21).

**Step 4: Find Q1 and Q3**

Q1 is the median (the middle) of the lower half of the data, and Q3 is the median (the middle) of the upper half of the data.

(3, 5, 7, 8, 9), | (11, 15, 16, 20, 21)

Q1 = 7 and Q3 = 16.


**Step 5: Subtract Q1 from Q3.**

16 – 7 = 9.
This is your IQR.

--------------------------------------------------------
*What if I Have an Odd Set of Numbers?*

*Example question: Find the IQR for the following data*

**Step 1: Put the numbers in order.**

1, 2, 5, 6, 7, 9, 12, 15, 18, 19, 27.

**Step 2: Find the median.**

1, 2, 5, 6, 7, 9, 12, 15, 18, 19, 27.

**Step 3: Place parentheses around the numbers above and below the median.**

Not necessary statistically, but it makes Q1 and Q3 easier to spot.

(1, 2, 5, 6, 7), 9, (12, 15, 18, 19, 27).

**Step 4: Find Q1 and Q3**

Think of Q1 as a median in the lower half of the data and think of Q3 as a median for the upper half of data.

(1, 2, 5, 6, 7),  9, ( 12, 15, 18, 19, 27). Q1 = 5 and Q3 = 18.

**Step 5: Subtract Q1 from Q3 to find the interquartile range.**

18 – 5 = 13.

**Standard Deviation**

- The Standard Deviation is a measure of how spread out numbers are. Symbol is σ (the greek letter sigma)

- The formula is easy: it is the square root of the Variance. 

**Variance**

- The average of the squared differences from the Mean.

**To calculate the variance follow these steps:**

1. Work out the Mean (the simple average of the numbers)
2. Then for each number: subtract the Mean and square the result (the squared difference).
3. Then work out the average of those squared differences. (Why Square?)

**EXAMPLE : 
The heights are: 10, 20, 30, 40 and 50. Find out the Mean, the Variance, and the Standard Deviation.**

*Mean =*	(10+20+30+40+50)/5 = 150/5 = 30m

*Variance =*

If we don't square, then result is 0

[(30-10)+(30-20)+(30-30)+(30-40)+(30-50)]/5 = [20+10+0-10-20]/5 = 0

That's why we square

(30-10)^2+(30-20)^2+(30-30)^2+(30-40)^2+(30-50)^2/5 = (400 + 100 + 0 + 100 + 400 )/5 = 1000/5 = 200

*Standard Deviation =*

Sqaure Root(500) = 22.36 (Same scale of the data)

**10. STANDARD DEVIATION**
- Standard Deviation = Square root(Variance) 
- SMALL SD means the values in the dataset are close to mean of the dataset on average
- LARGE SD means that the value in the dataset are far away from the mean of dataset on average
- In short, Standard Deviation measures how concentrated the data is around the mean
- More Concentration -> Smaller Standard Deviation

**A SMALL SD can be the goal in certain situations where results are restricted**
- Example, In Product Manufacutring and Quality Control companies. Suppose a particular type of car part that has to be 2 cm in diameter to properly fit cannot have a very BIG SD in manufacturing process. A BIG SD would mean most material would end up in trash because they don't fight right

**A HIGH SD reflects large amount of variation in the group**
- Example, if we look at the salaries for everyone in the company from intern to CEO, SD may be very large. On other hand if we see or observe only intern group, SD of salary may be low

**Inference :**
1. SD can be low or high depend on the scenario
2. OUTLIERS may affect the SD as the formula contains the mean
3. SD cannot be negative and lowest possible value is 0
4. 0 is only possible when every single entity is having same number (No Deviation)
5. SD have the same unit as original data

```
# 10) Standard Deviation
print("Standard Deivation for Boston Dataset is")
boston_pd.std()
```

**11. COVAIRANCE**

- Covariance is a statistical tool that is used to determine the relationship between the movement of two variables.

- When two stocks tend to move together, they are seen as having a positive covariance; when they move inversely, the covariance is negative.

- Covariance tells us how columns are related to each other.
  
  1. *If its positive value,* they are positively related. Coefficient/Value is positive means both variable tend to increase/decrease together
  2. *Coefficient/Value is negative* means if one variable increase other variable will decrease or vice-versa. 
  3. If it is close to 0, they are unrelated

**Problem -** It give us sign (+ or -) but not the strength i.e, how much they are related.  For Strength of relation, USE CORRELATION COEFFICIENT


```
# 11) Covariance
iris_pd.cov() 

# Inference :
# Try to remove one variable, if both the variables are highly correlated
```

**12. Correlation Coefficient**

  - Covariance find the direction of relationship but there is no range set for strength of relationship
  - Correlation find the exact value of strength in the relationship and the direction as well
  - *Range is from -1 to +1*
  - *+ve value suggest* they are positively correlated
  - *-ve value suggest* they are negatively correlated

**2 Methods are use to calculate Correlation Coefficient** 
  1. *Pearson correlation coefficient:* Assume data follows Normal Distribution and works well with linear data
  2. *Spearman correlation coefficient:* Don't assume any distribution. Works well with both linear & non-linear

By default while calculating, pearson method is used

**13. SKEWNESS**

  - In the curve of a distribution, the data on the right side of the curve may taper differently from the data on the left side. These taperings are known as "tails." 
  - *Negative skew refers* to a longer or fatter tail on the left side of the distribution, while *positive skew refers* to a longer or fatter tail on the right.
  - The mean of positively skewed data will be greater than the median. 
  - In a distribution that is negatively skewed, the exact opposite is the case: the mean of negatively skewed data will be less than the median. 
  - If the data graphs symmetrically, the distribution has zero skewness, regardless of how long or fat the tails are.
  - If *Skewness = 0,* data is perfectly symmetrical
  - If *Skewness is less than -1 or greater +1,* distribution is highly skewed
  - If *Skewness is between -1 and -0.5 or between 0.5 and 1,* distribution is moderately skewed
  - If *Skewness is between -0.5 and 0.5* then it is approximately symmetric  

```
# 13) Skewness
iris_pd.skew()
```

```
# For Boston dataset
boston_pd.skew()
```

```
# Randomly take one variable
print("Skewness of 'B' Column is",boston_pd['B'].skew()) 

# Inference : -2.89 which is highly skewed. Use Transformation to remove the skewness
```

```
# Plot the "B" variable
sns.distplot(boston_pd['B'])

# Inference - Left Skewed
```

```
# Check for another one variable
print("Skewness of 'CRIM' Column is ",boston_pd['CRIM'].skew())

# Inference : 5.22 which is highly skewed. 
# Use transformation - 1. Log  2. Square Root  3. Box Cox  
```

```
# Plot the Skewness
sns.distplot(boston_pd['CRIM'])

# Right Skewed 
```

```
# Log Transformation
import numpy as np
CRIM_Log = np.log(boston_pd['CRIM'])
print('After LOG TRANSFORMATION, Skewness is :',CRIM_Log.skew())
sns.distplot(CRIM_Log)
```

```
# Square Root Transformation
CRIM_sqrt = np.sqrt(boston_pd['CRIM'])
print('After Square Root Transformation, Skewness is ',CRIM_sqrt.skew())
sns.distplot(CRIM_sqrt)
```

```
# Box Cox Transformation
from scipy import stats
CRIM_BoxCox = stats.boxcox(boston_pd['CRIM'])[0]
Variable = pd.Series(CRIM_BoxCox)

print("After BoxCox Transformation, Skewness is ",Variable.skew())
sns.distplot(Variable)

# Inference :
# So the best one is BoxCox Transformation because of LOWEST SKEWNESS out of all three
```

```
# All the Skewness Value Calculated
print("Comparing all Skewness values")
pd.DataFrame([boston_pd['CRIM'].skew(),CRIM_Log.skew(),CRIM_sqrt.skew(),Variable.skew()],
             index=['Normal','Log','Square Root','Box Cox'],
             columns = ["Skewness"])

# Inference : Best transformations can be Log or Box Cox Transformation 
```

**14. KURTOSIS**

- Kurtosis is a statistical measure used to describe the degree to which scores cluster in the tails or the peak of a frequency distribution. 
- The peak is the tallest part of the distribution, and the tails are the ends of the distribution.
- There are three types of kurtosis: mesokurtic, leptokurtic, and platykurtic.
    1. *Mesokurtic:* Distributions that are moderate in breadth and curves with a medium peaked height.
    2. *Leptokurtic:* More values in the distribution tails and more values close to the mean (i.e. sharply peaked with heavy tails)
    3. *Platykurtic:* Fewer values in the tails and fewer values close to the mean (i.e. the curve has a flat peak and has more dispersed scores with lighter tails).

**What kurtosis is normal?**

- When kurtosis is equal to 3, the distribution is mesokurtic.
- This means the kurtosis is the same as the normal distribution, it is mesokurtic (medium peak).
- The kurtosis of a mesokurtic distribution is neither high nor low, rather it is considered to be a baseline for the two other classifications.

**Valur of Kurtosis Suggests**
- Positive Value means lots of data on tails
- Negative Value means little data on tails
- Standard Normal Distribution has a Kurtosis of 3
  1. Kurtosis < 3 - Platykurtic
  2. Kurtosis = 3 - Mesokurtic
  3. Kurtosis > 3 - Lepokurtic

```
# 14) Kurtosis

print("Before Transformation, Kurtosis is : ",boston_pd['CRIM'].kurtosis())
print("After Log Transformation, Kurtosis : ",CRIM_Log.kurtosis())
print("After Square Root Transformation, Kurtosis : ",CRIM_sqrt.kurtosis())
print("After Box Cox Transformation, Kurtosis : ",Variable.kurtosis())

# So the best one is Log Transformation or Box Cox Transformation
```
